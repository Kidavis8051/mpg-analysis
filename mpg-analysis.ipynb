#### *Create a visual showing how fuel economy has changed over time*

```python
# Use seaborn's theme for better visuals
sns.set_theme(style="whitegrid")

# Create a line plot: model_year vs avg mpg
sns.lineplot(data=mpg, x='model_year', y='mpg', estimator='mean', errorbar=None)

# Add titles and labels
plt.title('Fuel Economy Over Time')
plt.xlabel('Model Year')
plt.ylabel('Average Miles per Gallon (mpg)')

plt.show()
```

#### Fuel economy generally increased over time from 1970 to 1982.
#### Big jumps happened during the mid-late 1970s â€” possibly due to oil crises and stricter regulations.
#### Newer model years tend to have more efficient cars.

#### *Create two other visuals or tables related to fuel economy*

```python
# Boxplot of fuel economy by origin
sns.boxplot(data=mpg, x='origin', y='mpg')

# Add titles and labels
plt.title('Fuel Economy Distribution by Country of Origin')
plt.xlabel('Country of Origin')
plt.ylabel('Miles per Gallon (mpg)')

plt.show()
```

#### Japanese cars usually have higher mpg and less variation.
#### U.S. cars have a wider spread (some very low mpg vehicles).
#### European cars are in between.

```python
# Scatterplot of fuel economy vs weight
sns.scatterplot(data=mpg, x='weight', y='mpg')

# Add titles and labels
plt.title('Fuel Economy vs Vehicle Weight')
plt.xlabel('Vehicle Weight')
plt.ylabel('Miles per Gallon (mpg)')

plt.show()
```

#### Heavier vehicles tend to have lower fuel economy (negative correlation).
#### Lighter vehicles usually achieve better mpg.
#### There's a clear trend: lighter = more efficient.
